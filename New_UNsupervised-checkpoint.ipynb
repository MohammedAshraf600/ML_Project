```python
#Principal Component Analysis (PCA):

# Select features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
components = pca.fit_transform(X)
```

```python
# Create a new DataFrame with the principal components
df_pca = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
print(df_pca)
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('PCA - Principal Components Scatter Plot')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.grid(True)
plt.show()
```

```python
#Independent Component Analysis (ICA):

# Select features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]
```

```python
# Apply ICA for independent component analysis
ica = FastICA(n_components=3)
components = ica.fit_transform(X)
```

```python
# Create a new DataFrame with the independent components
df_ica = pd.DataFrame(data=components, columns=['IC1', 'IC2', 'IC3'])
print(df_ica)
```

```python
from mpl_toolkits.mplot3d import Axes3D


# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_ica['IC1'], df_ica['IC2'], df_ica['IC3'], alpha=0.5)
ax.set_title('ICA - Independent Components 3D Scatter Plot')
ax.set_xlabel('Independent Component 1 (IC1)')
ax.set_ylabel('Independent Component 2 (IC2)')
ax.set_zlabel('Independent Component 3 (IC3)')

plt.show()
```

```python
#DBSCAN:

#library for it
from sklearn.cluster import DBSCAN

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['cluster'] = dbscan.fit_predict(X)
print(df['cluster'])
```

```python
# Scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Pregnancies', y='Glucose', hue='cluster', data=df, palette='viridis', legend='full', alpha=0.7)
plt.title('DBSCAN Clustering')
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.show()
```

```python
#t-Distributed Stochastic Neighbor Embedding (t-SNE):

#library for it
from sklearn.manifold import TSNE

# Assuming df is your DataFrame with features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]


# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2)
components = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data=components, columns=['TSNE1', 'TSNE2'])
print(df_tsne)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Assuming df is your DataFrame with features
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[features]

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
components = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data=components, columns=['TSNE1', 'TSNE2'])

# Scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(df_tsne['TSNE1'], df_tsne['TSNE2'], alpha=0.7)
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE Component 1 (TSNE1)')
plt.ylabel('t-SNE Component 2 (TSNE2)')
plt.show()
```
