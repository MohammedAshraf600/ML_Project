```python
reshaped_images_train= np.expand_dims(reshaped_images_train,-1)
reshaped_images_test = np.expand_dims(reshaped_images_test,-1)
```

```python
#Feature Scaling
reshaped_images_train = reshaped_images_train/255
reshaped_images_test = reshaped_images_test/255
```

```python
#Spliting Dataset for validation set
reshaped_images_train, X_val , y_train , y_val = train_test_split(reshaped_images_train,y_train , test_size = 0.2 , random_state = 2020)
```

```python
reshaped_images_train.shape, y_train.shape  
```

```python
X_val.shape, y_val.shape
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(filters=64, kernel_size=3, strides=(1, 1), padding='valid', activation='relu', input_shape=[28, 28, 1]),  # 1st Layer
    keras.layers.MaxPooling2D(pool_size=(2, 2)),

    keras.layers.Conv2D(filters=128, kernel_size=3, strides=(2, 2), padding='same', activation='relu', input_shape=[28, 28, 1]),  # 2nd Layer
    keras.layers.MaxPooling2D(pool_size=(2, 2)),

    keras.layers.Conv2D(filters=64, kernel_size=3, strides=(2, 2), padding='same', activation='relu', input_shape=[28, 28, 1]),  # 3rd Layer
    keras.layers.MaxPooling2D(pool_size=(2, 2)),

    keras.layers.Flatten(),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(units=256, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(units=256, activation='relu'),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dropout(0.10),
    keras.layers.Dense(units=10, activation='softmax')  # Output layer
])
```

```python
keras.utils.plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model.summary()
```

```python
#Compile and Model Fitting
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(reshaped_images_train,y_train,epochs = 20,batch_size = 512,verbose = 1, validation_data=(X_val,y_val))
```

```python
#Testing the Model
model.predict(np.expand_dims(reshaped_images_test[0],axis = 0)).round(2)
# np.expand is used to change the 3 dimension data into 4 dimension
```

```python
np.argmax(model.predict(np.expand_dims(reshaped_images_test[0],axis = 0)).round(2))
```

```python
# cross check 
y_test.iloc[0]
```

```python
y_pred = model.predict(reshaped_images_test).round(2)
y_pred
```

```python
#Evaluation
model.evaluate(reshaped_images_test,y_test)
```

```python
plt.figure(figsize=(16, 16))
j = 1
for i in np.random.randint(0, len(y_test), 25):
    plt.subplot(5, 5, j)
    j += 1
    if i < len(reshaped_images_test):  # Check if the index is within the valid range
        plt.imshow(reshaped_images_test[i], cmap='Greys')
        plt.axis('off')
        predicted_label = class_labels[np.argmax(y_pred[i])]  # Assuming 'predictions' is your model's predictions
        plt.title('Actual = {} / {}\nPredicted = {} / {}'.format(class_names[y_test.iloc[i]], y_test.iloc[i],predicted_label, np.argmax(y_pred[i])))

plt.show()
```

```python
#Confusion Matrix
plt.figure(figsize = (16,9))
y_pred_labels = [np.argmax(label) for label in y_pred]
cm = confusion_matrix(y_test,y_pred_labels)

# HeatMap
sns.heatmap(cm , annot = True,fmt = 'd',xticklabels = class_labels,yticklabels = class_labels)
```

```python
cr = classification_report (y_test,y_pred_labels,target_names = class_labels)
print(cr)
```
